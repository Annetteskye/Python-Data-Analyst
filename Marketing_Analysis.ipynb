```python
import pandas as pd

# 1. Загружаем данные (имитируем выгрузку из БД)
data = {
    'Category': ['Dresses', 'Shoes', 'Accessories', 'Bags'],
    'Revenue': [1500000, 800000, 300000, 1200000],
    'Marketing_Cost': [200000, 150000, 30000, 180000],
    'COGS': [600000, 350000, 100000, 500000] # Себестоимость
}
df = pd.DataFrame(data)

# 2. Считаем Валовую прибыль (Gross Profit)
df['Profit'] = df['Revenue'] - df['COGS']

# 3. Считаем ROAS (Окупаемость рекламы) = Выручка / Расходы на маркетинг
df['ROAS'] = (df['Revenue'] / df['Marketing_Cost']).round(2)

# 4. Считаем Маржинальность (%)
df['Margin_Pct'] = ((df['Profit'] / df['Revenue']) * 100).round(1)

# Выводим только прибыльные категории с ROAS > 5
top_categories = df[df['ROAS'] > 5].sort_values(by='ROAS', ascending=False)
print(top_categories)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Данные (наш мини-кейс по категориям)
data = {
    'Category': ['Платья', 'Обувь', 'Аксессуары', 'Сумки'],
    'Revenue': [12000, 45000, 8000, 25000],
    'Marketing_Cost': [2000, 8000, 1500, 4000],
    'COGS': [5000, 20000, 3000, 10000]
}
df = pd.DataFrame(data)

# 2. Расчеты
df['Profit'] = df['Revenue'] - df['COGS']
df['ROAS'] = (df['Revenue'] / df['Marketing_Cost']).round(2)

# 3. Визуализация (График окупаемости рекламы)
plt.figure(figsize=(10, 5))
plt.bar(df['Category'], df['ROAS'], color='skyblue')
plt.axhline(y=5, color='red', linestyle='--', label='Target ROAS (5.0)') # Целевая линия
plt.title('Окупаемость рекламы (ROAS) по категориям', fontsize=14)
plt.ylabel('ROAS', fontsize=12)
plt.legend()
plt.show()
```
